the goal is to turn everything into text files first and then try to identify duplicates because the html structure is not very helpful.

```python
def process_html_file(input_file_path, output_file_path):
    with open(input_file_path, 'r', encoding='utf-8') as file:
        html_content = file.read()

    soup = BeautifulSoup(html_content, 'html.parser')

    # Extract all text from the HTML
    full_text = soup.get_text(separator='\n')

    # Split the text into sections based on multiple newlines
    sections = re.split(r'\n\s*\n', full_text)

    # Create structured output
    structured_text = "\n\n".join(sections)

    # Save to a text file
    with open(output_file_path, 'w', encoding='utf-8') as file:
        file.write(structured_text)

def process_folder(input_folder, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for filename in os.listdir(input_folder):
        if filename.endswith('.html'):
            input_file_path = os.path.join(input_folder, filename)
            output_file_path = os.path.join(output_folder, f"{os.path.splitext(filename)[0]}.txt")
            process_html_file(input_file_path, output_file_path)
            print(f"Processed {input_file_path} to {output_file_path}")

# Example usage
input_folder = '/Users/mervetekgurler/Desktop/PhD/CESTA/cesta-events/wayback_machine/wayback_html'
output_folder = '/Users/mervetekgurler/Desktop/PhD/CESTA/cesta-events/wayback_machine/raw_text'
process_folder(input_folder, output_folder)
```

I experimented with using chatgpt and getting JSON output from the model. It workes quite well, so I will use it in this case. 
I narrowed the dataset down to the stuff between 2012 and end of 2016 because that it the period we are missing.

```python
import os
import json
from dotenv import load_dotenv
from openai import OpenAI

# Load environment variables from .env file
load_dotenv()

# Get the OpenAI API key from the environment variable
api_key = os.getenv('OPENAI_API_KEY')

# Initialize the OpenAI client
client = OpenAI()

# Function to read the content of a file
def read_file(file_path):
    with open(file_path, 'r') as file:
        return file.read()

# Function to restructure data using OpenAI API
def restructure_data(file_content):
    completion = client.chat.completions.create(
        messages=[
            {
                "role": "user",
                "content": f"Restructure the following data into JSON format without inventing or changing any content. Only convert the structure:\n\n{file_content}"
            }
        ],
        model="gpt-4-turbo",
        response_format={"type": "json_object"}
    )
    return completion.choices[0].message.content
```

```python
# Read the content of the input file
input_file_path = '/Users/mervetekgurler/Desktop/PhD/CESTA/cesta-events/wayback_machine/2012-2016/20120705054801.txt'  # Replace with your input file path
file_content = read_file(input_file_path)

# Get the restructured data from OpenAI
restructured_data = restructure_data(file_content)

# Print the restructured data
print("Restructured Data:")
print(restructured_data)

# save the JSON data to a file
output_file_path = 'restructured_data.json'  # Replace with your desired output file path
with open(output_file_path, 'w') as file:
    file.write(restructured_data)

    print(f"Restructured data has been saved to {output_file_path}")
```

```python
output_folder = '/Users/mervetekgurler/Desktop/PhD/CESTA/cesta-events/wayback_machine/2012-2016_json'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

input_folder = '/Users/mervetekgurler/Desktop/PhD/CESTA/cesta-events/wayback_machine/2012-2016_texts'

# Loop over the files in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith('.txt'):
        input_file_path = os.path.join(input_folder, filename)
        output_file_path = os.path.join(output_folder, f"{os.path.splitext(filename)[0]}.json")
        
        file_content = read_file(input_file_path)
        # Call the restructure function on the file
        restructured_data = restructure_data(file_content)
        
        with open(output_file_path, 'w') as file:
            file.write(restructured_data)
        
        print(f"Processed {input_file_path} and saved to {output_file_path}")
```
