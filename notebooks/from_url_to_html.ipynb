```python
import os
import re
import pandas as pd
import csv
import numpy as np
import requests
```

```python
# Create a directory for HTML files if it doesn't exist
folder_name = '/Users/mervetekgurler/Desktop/PhD/CESTA/cesta-events/events/event_html'
if not os.path.exists(folder_name):
    os.makedirs(folder_name)

# Load the CSV file
df = pd.read_csv('/Users/mervetekgurler/Desktop/PhD/CESTA/cesta-events/events/webscraper_data.csv')


def download_html(url, file_name):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            file_path = os.path.join(folder_name, file_name)
            with open(file_path, 'w', encoding='utf-8') as file:
                file.write(response.text)
                print(f"Saved HTML for {url} as {file_path}")
        else:
            print(f"Failed to download {url}: Status code {response.status_code}")
    except Exception as e:
        print(f"Error downloading {url}: {e}")

# Loop through DataFrame
for index, row in df.iterrows():
    file_name = f"{row['web-scraper-order']}.html"
    download_html(row['pagelink2-href'], file_name)
```
